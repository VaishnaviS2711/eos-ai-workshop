```python
import numpy as np
import matplotlib.pyplot as plt

# 16x16 grid, 0 = black, 1 = yellow
face = np.zeros((16, 16), dtype=int)

# outline of face (circle shape)
face[0, 4:12] = 1
face[1, 2:14] = 1
face[2, 1:15] = 1
face[3, 1:15] = 1
face[4, 0:16] = 1
face[5, 0:16] = 1
face[6, 0:16] = 1
face[7, 0:16] = 1
face[8, 0:16] = 1
face[9, 0:16] = 1
face[10, 0:16] = 1
face[11, 0:16] = 1
face[12, 1:15] = 1
face[13, 1:15] = 1
face[14, 2:14] = 1
face[15, 4:12] = 1

# eyes
face[5, 4:6] = 0
face[5, 10:12] = 0
face[6, 4:6] = 0
face[6, 10:12] = 0

# smile
face[10, 3:5] = 0
face[10, 11:13] = 0
face[11, 4:6] = 0
face[11, 10:12] = 0
face[12, 5:11] = 0

# plot
plt.figure(figsize=(5,5))
plt.imshow(face, cmap='YlOrBr')
plt.axis('off')
plt.title("Smiley Face :)")
plt.show()
```
